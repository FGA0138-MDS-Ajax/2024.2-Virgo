```python
import tensorflow as tf
import matplotlib.pyplot as plot

# Clearing some garbage that may be in memory
tf.keras.backend.clear_session()
IMAGE_SIZE = 256
BATCH_SIZE = 16 #32 -> 16
EPOCH= 32
#this path would be used had we downloaded the dataset from tfds:
#   "/root/tensorflow_datasets/downloads/extracted/ZIP.data.mend.com_publ-file_data_tywb_file_d565-c1rDQyRTmE0CqGGXmH53WlQp0NWefMfDW89aj1A0m5D_A/Plant_leave_diseases_dataset_without_augmentation"
#good thing we did not!
base_data = tf.keras.preprocessing.image_dataset_from_directory("le_dataset",
                                                                shuffle=True,
                                                                image_size=(IMAGE_SIZE,IMAGE_SIZE),
                                                                batch_size=BATCH_SIZE)

print(base_data)
```

```python
# Here we are taking the batches to separate the data accordingly
# We may or may not want to randomize it
val_batches = int(total_batches * 0.2)
train_batches = total_batches - val_batches
train_data = base_data.skip(int(val_batches))  # Skip the validation batches for training
val_data = base_data.take(int(val_batches))   # Take the validation batches

print(train_data)
```

# STEP 5: Normalization and Patterns
Here we define a pattern using tf.keras.layers.Rescaling

```python
#takes about a minute to run.

data_augmentation = tf.keras.Sequential(
    [
        tf.keras.layers.RandomFlip("horizontal"),
        tf.keras.layers.RandomRotation(0.1),
        tf.keras.layers.RandomZoom(0.1),
    ]
)

data_resizing = tf.keras.Sequential(
    [
        tf.keras.layers.Resizing(width = IMAGE_SIZE,
                                 height = IMAGE_SIZE,
                                 interpolation = "bilinear"
                                 ),
        tf.keras.layers.Rescaling(1./255.)
    ]
)

def normalize(image,label):
  image = data_augmentation(image)
  image = data_resizing(image)
  image = tf.cast(image, tf.float32)
  return image, label
# Basically there are 2 ways to use this
# Gonna try this one first
normalized_data = train_data.map( normalize,
                                  num_parallel_calls = tf.data.AUTOTUNE
                                  )
#seems to be working fine.
#image_batch, labels_batch = next(iter(normalized_data))
#first_image = image_batch[0]
#first_label = labels_batch[0]
# Try to configure it to PERFORM
AUTOTUNE = tf.data.AUTOTUNE
#normalized_data = normalized_data.cache().prefetch(buffer_size=AUTOTUNE)
train_data = train_data.cache().prefetch(buffer_size=AUTOTUNE)
val_data = val_data.cache().prefetch(buffer_size=AUTOTUNE)

# This can be useful later on
# https://www.tensorflow.org/guide/data_performance?hl=pt-br
# This could work way better IF the presentation wasnt in 13h and 5 min
```

```python
#this is looking somewhat useful, so i'll try to use it.
import gc

class gc_class(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        gc.collect()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001,
                                     beta_1=0.9,
                                     beta_2=0.999,
                                     epsilon=0.1)

use_less_ram_please = gc_class()

#this mikefoxtrot seems to be the root of our trouble :)
our_metrics = [
    tf.keras.metrics.Accuracy(name='accuracy'),
    tf.keras.metrics.Precision(name='precision'),
    tf.keras.metrics.Recall(name='recall'),
    tf.keras.metrics.AUC(name='auc')
]

# Compiling
model.compile( optimizer = optimizer,
               loss      = 'sparse_categorical_crossentropy',
               metrics   = ['accuracy']
               )

hist = model.fit(train_data,
                 epochs=EPOCH,
                 steps_per_epoch = 150,
                 validation_data=val_data,
                 validation_steps=val_batches,
                 callbacks = [
                     tf.keras.callbacks.EarlyStopping(monitor='accuracy'),
                     use_less_ram_please()
                     ]
                 ) # This better not nuke everything #it might. #it did not.
hist.history
```
